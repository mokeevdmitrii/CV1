```python
import numpy as np
import cv2
import typing as tp
import os

import matplotlib.pyplot as plt

print(f'Numpy version: {np.__version__}')
print(f'opencv-python version: {cv2.__version__}')
```

```python
# out = (1 - w) * x + w * background,     
class Mixture:
    def __init__(self, back_weight=0.5):
        self.back_weight = 0.5
        
    def __call__(self, x, back):
        back = cv2.resize(back, (x.shape[1], x.shape[0]))
        res = cv2.addWeighted(x, 1. - self.back_weight, back, self.back_weight, 0)
        return res
    
class MixtureAlpha:
    def __call__(self, x, back):
        
        colors, alpha_float = x[:, :, :3], (x[:, :, 3] / 255.)[..., None]
        out = colors * alpha_float + back * (1. - alpha_float)
        return out.astype(np.uint8)
```

```python
# https://pytorch.org/vision/main/generated/torchvision.transforms.RandomPerspective.html
class RandomPerspective:
    
    def __init__(self, distortion_scale=0.25, gen=np.random.default_rng()):
        self.distortion_scale = distortion_scale
        self.gen = gen
        
    def __call__(self, x: cv2.Mat) -> cv2.Mat:
        
        h, w = x.shape[:2]
        half_height, half_width = h // 2, w // 2
        
        top_left = [self.gen.integers(0, int(self.distortion_scale * half_width)) + 1, 
                    self.gen.integers(0, int(self.distortion_scale * half_height) + 1)]
        
        top_right = [self.gen.integers(w - int(self.distortion_scale * half_width) - 1, w),
                     self.gen.integers(0, int(self.distortion_scale * half_height) + 1)]
        
        bottom_left = [self.gen.integers(0, int(self.distortion_scale * half_width)) + 1,
                       self.gen.integers(h - int(self.distortion_scale * half_height) - 1, h)]
        
        bottom_right = [self.gen.integers(w - int(self.distortion_scale * half_width) - 1, w),
                       self.gen.integers(h - int(self.distortion_scale * half_height) - 1, h)]
        
        from_points = np.array([[0, 0], [w, 0], [0, h], [w, h]], dtype=np.float32)
        to_points = np.array([top_left, top_right, bottom_left, bottom_right], dtype=np.float32)
        
        conv_matrix = cv2.getPerspectiveTransform(from_points, to_points)
        
        warped_image = cv2.warpPerspective(x, conv_matrix, (w, h))
        # maybe crop right here (by self.distortion_scale)
        return warped_image
```

```python
# my favorite - affine transforms
class RandomRotate:
    
    def __init__(self, angle: int, fill_value: int= 0, gen=np.random.default_rng()):
        self.angle = angle
        self.fill_value = fill_value
        self.gen = gen

    def __call__(self, img: cv2.Mat) -> cv2.Mat:
        h, w = img.shape[:2]
        num_ch = img.shape[2]
        angle = self.gen.uniform(-self.angle, self.angle)
        rotate_matrix = cv2.getRotationMatrix2D((w / 2, h / 2), angle, 1)
        img = cv2.warpAffine(img, rotate_matrix, (w, h), 
                             borderMode=cv2.BORDER_CONSTANT, borderValue=(self.fill_value, ) * num_ch)
        return img

    
class RandomTranslate:
    
    def __init__(self, shift: float, fill_value: int = 0, gen=np.random.default_rng()):
        self.shift = shift
        self.fill_value = fill_value
        self.gen = gen
        
    def __call__(self, img: cv2.Mat) -> cv2.Mat:
        h, w = img.shape[:2]
        num_ch = img.shape[2]
        shift_x, shift_y = (self.gen.uniform(-self.shift, self.shift) * w, 
                            self.gen.uniform(-self.shift, self.shift) * h)
        shift_matrix = np.array([[1, 0, shift_x], [0, 1, shift_y]])
        img = cv2.warpAffine(img, shift_matrix, (w, h),
                             borderMode=cv2.BORDER_CONSTANT, borderValue=(self.fill_value, ) * num_ch)
        
        return img
        
        
class RandomResize:
    
    def __init__(self, resize: tp.Tuple[float, float], pad: bool, fill_value: int = 0, gen=np.random.default_rng()):
        self.min_scale, self.max_scale = resize
        self.fill_value = fill_value
        self.pad = pad
        self.gen = gen
        
    def __call__(self, img: cv2.Mat) -> cv2.Mat:
        h, w, _ = img.shape
        scale = self.gen.uniform(self.min_scale, self.max_scale)

        resized = cv2.resize(img, (int(scale * w), int(scale * h)))
        if not self.pad:
            return resized
        
        h_r, w_r, _ = resized.shape
        out = np.ones_like(img) * self.fill_value
        y_s, x_s = (h - h_r) // 2, (w - w_r) // 2
        out[y_s:y_s+h_r, x_s:x_s+w_r] = resized
        
        return out
        
```

```python
class RandomAffine:
    
    def __init__(self, angle: tp.Optional[int] = None, 
                       shift: tp.Optional[float] = None,
                       resize: tp.Optional[tp.Tuple[float, float]] = None,
                       fill_value: int = 0, gen = np.random.default_rng()):
        
        self.rotate = RandomRotate(angle, fill_value, gen) if angle is not None else None
        self.translate = RandomTranslate(shift, fill_value, gen) if shift is not None else None  
        self.resize = RandomResize(resize, True, fill_value, gen) if resize is not None else None
        
    def __call__(self, img):
        
        if self.resize is not None:
            img = self.resize(img)
        
        if self.rotate is not None:
            img = self.rotate(img)
        
        if self.translate is not None:
            img = self.translate(img)
        
        return img
    
```

```python
class RandomSharpness:
    
    def __init__(self, sharpness: tp.Tuple[float, float], 
                       sigma: tp.Tuple[float, float], gen=np.random.default_rng()):
        self.sharpness = sharpness
        self.sigma = sigma 
        self.gen = gen
        
    def __call__(self, img):
        sigma = self.gen.uniform(self.sigma[0], self.sigma[1])
        sharpness = self.gen.uniform(self.sharpness[0], self.sharpness[1])
        
        blurred = cv2.GaussianBlur(img, (0, 0), sigma)
        result = sharpness * (img - blurred) + img
        cv2.normalize(result, result, 0, 255, cv2.NORM_MINMAX, dtype=-1)
        result = result.astype(np.uint8)
        return result
        
```

```python
class NoiseNBlur:
    
    def __init__(self, noise_sigma: float, sigma: tp.Tuple[float, float], gen=np.random.default_rng()):
        
        self.sigma_min, self.sigma_max = sigma
        self.noise_sigma = noise_sigma
        self.gen = gen
        
    def __call__(self, img):
        h, w, _ = img.shape
        
        sigma = self.gen.uniform(self.sigma_min, self.sigma_max)
        result = cv2.GaussianBlur(img, (0, 0), sigma)
        
        gauss_noise = self.gen.normal(0, self.noise_sigma, (h, w, 1))
        result = result + gauss_noise
        
        cv2.normalize(result, result, 0, 255, cv2.NORM_MINMAX, dtype=-1)
        result = result.astype(np.uint8)

        return result
        
```

```python
class ProbableTransform:
    def __init__(self, callble, p=0.5, gen=np.random.default_rng()):
        self.callble = callble
        self.p = p
        self.gen = gen
        
    def __call__(self, img):
        if self.gen.random() > self.p:
            return img
        return self.callble(img)
        
```

```python
class Sequential:
    def __init__(self, callables):
        self.callables = callables

    def __call__(self, img):
        out = img
        for c in self.callables:
            out = c(out)
        return out
```

```python
IMAGE_FOLDER = './input/images'
BACK_FOLDER = './input/background'
PAPER_FOLDER = './input/destroyed_papers'
```

```python
def is_jpg(file: str):
    return file.endswith('.jpg')

def read_image(file: str):
    img = cv2.imread(file)
    if img is None:
        raise ValueError(f"No valid image by path: {file}")
    return img

def select_random_pic(directory, gen=np.random.default_rng()):
    images = list(filter(is_jpg, os.listdir(directory)))
    return read_image(os.path.join(directory, gen.choice(images)))

def get_random_paper():
    return select_random_pic(PAPER_FOLDER)

def get_random_background():
    return select_random_pic(BACK_FOLDER)
```

```python
def convert_to_alpha(img):
    if img.shape[-1] == 3:
        return cv2.cvtColor(img, cv2.COLOR_BGR2BGRA)
    elif img.shape[-1] == 4:
        return result
    else:
        raise ValueError(f"invalid image shape: {img.shape}")
```

```python
def add_random_background(image):
    background = get_random_background()
    
    # resize to match smallest background dimension
    bh, bw, _ = background.shape
    h, w, _ = image.shape
    min_factor = min((bh / h, bw / w))
    image = cv2.resize(image, (int(min_factor * w), int(min_factor * h)))
    
    out = RandomResize((0.5, 0.85), pad=False)(image)
    h, w, _ = out.shape

    # maybe Resize, but to fixed size
    image = np.zeros((bh, bw, 4), dtype=np.uint8)
    y_s, x_s = (bh - h) // 2, (bw - w) // 2
    image[y_s:y_s+h, x_s:x_s+w] = convert_to_alpha(out)

    image = RandomAffine(angle=15, shift=0.2)(image)
    image = MixtureAlpha()(image, background)
    return image
    
    
def add_random_paper(image):
    paper = get_random_paper()
    return Mixture(0.5)(image, paper)
```

```python
AFFINES = Sequential((
    RandomResize((0.8, 0.9),pad=True,fill_value=255),
    RandomRotate(5,fill_value=255),
    RandomTranslate(0.1,fill_value=255),
))

ADD_PAPER = ProbableTransform(add_random_paper, p=1.0)
ADD_BACKGROUND = ProbableTransform(add_random_background, p=0.8)

FINALISAMOS = Sequential((
    NoiseNBlur(noise_sigma=10, sigma=(0.05, 0.5)),
    ProbableTransform(RandomPerspective(distortion_scale=0.25), p=0.8),
    ProbableTransform(RandomSharpness(sharpness=(0.01, 0.1), sigma=(1.0,2.0)), p=0.8)
))
```

```python
def augment_image(image):
    
    out = AFFINES(image)
    
    out = ADD_PAPER(out)
    out = ADD_BACKGROUND(out)
    out = FINALISAMOS(out)
    
    return out
```
