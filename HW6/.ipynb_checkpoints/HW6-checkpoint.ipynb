```python
def read_folder(folder: str) -> list[np.ndarray]:
    imgs = []
    imgs_paths = os.listdir(folder)
    for im_path in imgs_paths:
        path = os.path.join(folder, im_path)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        imgs.append(img)
    return imgs
```

```python
def get_bboxs(img):
    qr_detector = cv2.QRCodeDetector()
    has_found, points = qr_detector.detectMulti(img.copy())
    if has_found:
        return points
    return []
```

```python
MAPPED_SIZE = 400

def warp_qr_perspective(img, bbox):
    
    qr_points = np.array([[0,0],[MAPPED_SIZE,0],[MAPPED_SIZE,MAPPED_SIZE],[0,MAPPED_SIZE]],dtype=np.float32)
    transform = cv2.getPerspectiveTransform(bbox, qr_points)
    _, inverse_matrix = cv2.invert(transform)
    #     ,   
    res = cv2.warpPerspective(img, transform, (MAPPED_SIZE, MAPPED_SIZE))
    return res, inverse_matrix
    
```

```python
img = TEST_1[0]
bbox = get_bboxs(img)[0]
qr_img, _ = warp_qr_perspective(img, bbox)
```

```python
def add_markup(img, bbox, thickness):
    bbox = np.array(bbox).astype(int)
    if len(img.shape) == 2:
        rgb = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    else:
        rgb = img
    res = cv2.rectangle(rgb, bbox[0], bbox[1], (255, 0, 0), thickness)
    return res
```

```python
def find_clean_markup():
    return [
        [[0, 0], [65, 65]],
        [[0, 335], [65, 400]],
        [[335, 0], [400, 65]]
    ]

def add_markup_to_clean(img):
    bboxs = find_clean_markup()
    res = img.copy()
    for b in bboxs:
        res = add_markup(res, b, thickness=5)
    return res
```

```python
def map_point(matrix, p):
    x, y, z = matrix @ np.array([p[0], p[1], 1], dtype=np.float32)
    return [x / z, y / z]

def map_bbox(matrix, bbox):
    return map_point(matrix, bbox[0]), map_point(matrix, bbox[1])
    
```

      .    

```python
def evaluate(img, show=True):
    
    maybe_qrs = get_bboxs(img)
    
    result = img.copy()
    
    for qr in maybe_qrs:
        _, inverse_m = warp_qr_perspective(img, qr)
        clean_bboxes = find_clean_markup()
        bboxes_orig = [map_bbox(inverse_m, bbox) for bbox in clean_bboxes]
        
        for bbox in bboxes_orig:
            result = add_markup(result, bbox, thickness=5)
    if show:
        plt.imshow(result)
    
    return result
    
```
